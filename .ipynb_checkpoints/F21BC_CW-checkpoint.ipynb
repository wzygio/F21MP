# Task1-Implement a multi-layer ANN architecture

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)
```

We define a class "NeuralNetwork" to represent the entire network, including initialization, weight and bias generation, and forward propagation functions.

Parameter list:  
**input_size**: the number of neurons in the input layer  
**hidden_layers**: a list of the number of neurons in the hidden layer (such as [10, 20, 10] for three hidden layers)  
**output_size**: the number of neurons in the output layer  
**activation_function**: the activation function used, optional 'logistic', 'relu', 'tanh'

```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_layers, output_size, activation_function='relu'):
        # Network layer structure
        self.layers = [input_size] + hidden_layers + [output_size]
        self.activation_function = self.get_activation_function(activation_function)
        
        # Initialize weights and biases
        self.weights = []  
        self.biases = []   
        
        for i in range(len(self.layers) - 1):
            # Xavier Initialization for weights
            n_in = self.layers[i]
            n_out = self.layers[i + 1]
            # Xavier normal distribution initialization
            self.weights.append(np.random.randn(n_in, n_out) * np.sqrt(1 / n_in))
            # Initialize bias to zero
            self.biases.append(np.zeros((1, n_out)))

    def get_activation_function(self, name):
        if name == "sigmoid":
            return sigmoid
        elif name == "relu":
            return relu
        elif name == "tanh":
            return tanh
        else:
            raise ValueError("Unknown activation function")

    def forward(self, X):
        """
        Forward propagation calculation
        X: Input data, shape is (number of samples, number of features)
        Return value: Output result, shape is (number of samples, number of neurons in the output layer)
        """
        A = X  # Input layer output
        for i in range(len(self.weights) - 1):
            Z = np.dot(A, self.weights[i]) + self.biases[i]  # Linear transformation
            A = self.activation_function(Z)  # Activation function
        # Output layer has no activation function
        output = np.dot(A, self.weights[-1]) + self.biases[-1]
        return output
    
    def to_particle(self):
        """
        Flatten the weights and biases into a one-dimensional array for particle representation
        :return: Flattened one-dimensional array
        """
        flattened_weights = [w.flatten() for w in self.weights]
        flattened_biases = [b.flatten() for b in self.biases]
        return np.concatenate(flattened_weights + flattened_biases)

    def from_particle(self, particle):
        """
        Restore the one-dimensional array of particles to their original form of weights and biases
        :param particle: the flattened one-dimensional array
        """
        start = 0
        for i in range(len(self.weights)):
            # Restore weights
            weight_shape = self.weights[i].shape
            weight_size = np.prod(weight_shape)
            self.weights[i] = particle[start:start + weight_size].reshape(weight_shape)
            start += weight_size
            
            # Restore bias
            bias_shape = self.biases[i].shape
            bias_size = np.prod(bias_shape)
            self.biases[i] = particle[start:start + bias_size].reshape(bias_shape)
            start += bias_size

    def get_bounds(self, bound_value=1):
        """
        Generate the search space range according to the structure of the neural network
        :param bound_value: The upper and lower bounds of the search space, for example [-bound_value, bound_value]
        :return: NumPy array, shape (dimensions, 2), each row represents [min, max]
        """
        particle_size = self.to_particle().size  # 获取粒子的总维度
        bounds = np.array([[-bound_value, bound_value]] * particle_size)  # 为每个维度设置相同的上下界
        return bounds
```

# Task2-Implement a PSO algorithm

Each particle represents a solution (ANN weights and biases), containing the following information:  
Position: represents the current solution.  
Velocity: represents the direction of change of the solution.  
Personal optimal solution (pbest): the best position of the particle in history.  
Fitness: performance evaluation of the particle's current position.

```python
class Particle:
    def __init__(self, dimensions, bounds):
        """
        Initialize particles
        :param dimensions: the dimensions of the particle position (corresponding to the total number of weights and biases of the ANN)
        :param bounds: the search space range [(min, max), ...] the minimum and maximum values ​​of each dimension
        """
        bounds_range = bounds[:, 1] - bounds[:, 0]
        self.position = np.random.uniform(-0.5 * bounds_range, 0.5 * bounds_range, dimensions)  # Randomly initialize the position
        self.velocity = np.random.uniform(-0.2 * bounds_range , 0.2 * bounds_range , dimensions)  # Randomly initialize speed
        self.pbest_position = self.position.copy()  # Initialize the optimal position of the individual
        self.pbest_value = float('inf')  # Initialize the individual optimal value to infinity
        self.fitness = float('inf')  # Current fitness value
```

The Swarm class is a particle swarm class that stores multiple particles and is responsible for updating the Informants mechanism. After discussion, we believe that Informants is the biggest factor affecting the convergence effect, because it determines **l_best**, and **l_best** determines the exploration direction of particles in the search space.

However, in order to balance the convergence speed and convergence accuracy, we designed a set of strategies for dynamically adjusting Informants. Specifically, we divide PSO into three stages:

Initial (exploration stage): Randomly assign Informants to increase the diversity of search.  
Mid-term (balanced exploration and development stage): Select Informants based on distance to enhance local relevance.  
Late (development stage): Select Informants based on fitness to accelerate convergence.  

Based on the above mechanism, we designed the **update_informants**() function to be responsible for the update of Informants.

```python
class Swarm:
    def __init__(self, swarm_size, dimensions, bounds, num_informants):
        """
        Initialize particle swarm
        :param swarm_size: particle swarm size
        :param dimensions: position dimension of each particle
        :param bounds: search space range [(min, max), ...]
        :param num_informants: number of informants for each particle
        """
        self.particles = [Particle(dimensions, bounds) for _ in range(swarm_size)]
        self.gbest_position = np.zeros(dimensions)  # Initialize to zero vector
        self.gbest_value = float('inf')  # Group optimal value
        
        # Informants allocation
        self.informants = self.assign_informants(swarm_size, num_informants)

    def assign_informants(self, swarm_size, num_informants):
        """
        Randomly assign Informants (particle objects)
        :param num_informants: Number of Informants for each particle
        :return: Informants assignment list (a list of Informants objects corresponding to each particle)
        """
        informants = []
        for i in range(swarm_size):
            informant_indices = random.sample(range(swarm_size), max(1, num_informants))
            informants.append(informant_indices)
        return informants

    def update_informants(self, num_informants, m):
        """
        Dynamically update the informants of each particle, combining random assignment, distance and fitness strategies
        """
        swarm_size = len(self.particles)
        exploration_ratio = 1 - m  # Current exploration ratio
    
        for i in range(swarm_size):
            if exploration_ratio > 0.5:  # Initial: Random Assignment
                self.informants[i] = random.sample(range(swarm_size), num_informants)
            elif exploration_ratio > 0.2:  # Medium term: distance-based
                distances = [
                    (j, np.linalg.norm(self.particles[i].position - self.particles[j].position))
                    for j in range(swarm_size) if j != i
                ]
                sorted_indices = [j for j, _ in sorted(distances, key=lambda x: x[1])]
                self.informants[i] = sorted_indices[:num_informants]
            else:  # Late stage: based on fitness
                sorted_indices = sorted(range(swarm_size), key=lambda x: self.particles[x].fitness)
                self.informants[i] = sorted_indices[:num_informants]
```

The fitness function is used to evaluate the quality of the solution of the current particle. As required, the fitness value is set to the prediction error (mse) of the ANN. In addition, we introduce a regularization term to avoid overfitting.

```python
def fitness_function(particle_position, ann, X_train, y_train, lambda_reg=0.01):
    """
    Calculate fitness value (add L2 regularization)
    :param particle_position: particle position (expanded weight and bias array)
    :param ann: neural network model
    :param X_train: training set input
    :param y_train: training set output
    :param lambda_reg: regularization coefficient
    :return: fitness value (loss + regularization term)
    """
    # Decode the particle's position into weights and biases
    ann.from_particle(particle_position)

    # Forward propagation, calculate the predicted value
    predictions = ann.forward(X_train)

    # Calculate mean square error (MSE)
    mse = np.mean((predictions - y_train) ** 2)

    # Calculate L2 regularization term
    l2_regularization = 0
    for w in ann.weights:
        l2_regularization += np.sum(w ** 2)

    # Total loss = MSE + regularization term
    total_loss = mse + lambda_reg * l2_regularization
    return total_loss
    # return np.log(1 + mse) 
```

Update the particle's velocity and position according to the PSO formula. Here, we introduce the use of the **rebound mechanism** for boundary processing.

```python
def update_particle(particle, lbest_position, w, c1, c2, bounds_range, m):
    """
    Update the position and velocity of particles
    :param particle: current particle
    :param lbest_position: Informants optimal position
    :param w: inertia weight
    :param c1: cognitive factor
    :param c2: social factor
    :bounds: search space
    :max_velocity: limit the maximum velocity of particles
    """
    r1, r2 = np.random.rand(), np.random.rand()
    
    # Speed update formula
    particle.velocity = (
        w * particle.velocity +
        c1 * r1 * (particle.pbest_position - particle.position) +
        c2 * r2 * (lbest_position - particle.position)
    )

    # Dynamically limit the speed range of particles so that the maximum speed of particles decreases as the number of iterations increases
    # max_velocity = 0.2 * (1 - (m) ** 2)
    # particle.velocity = np.clip(particle.velocity, -max_velocity, max_velocity)
    
    # Position update formula
    particle.position += particle.velocity  
    
    # Boundary processing: using a bounce mechanism
    particle.position = apply_bounds_with_reflection(particle.position, bounds)

def apply_bounds_with_clip(position, bounds):
    """
    Use the rebound mechanism to limit the particle position within the search space
    :param position: particle position array
    :param bounds: search space range, shape is (dimensions, 2)
    :return: limited particle position
    """
    position = np.clip(position, bounds[:, 0], bounds[:, 1])
    return position

def apply_bounds_with_reflection(position, bounds):
    """
    Use the rebound mechanism to limit the particle position within the search space
    :param position: particle position array
    :param bounds: search space range, shape is (dimensions, 2)
    :return: limited particle position
    """
    for i in range(len(position)):
        if position[i] < bounds[i, 0]:
            position[i] = bounds[i, 0] + (bounds[i, 0] - position[i])
        elif position[i] > bounds[i, 1]:
            position[i] = bounds[i, 1] - (position[i] - bounds[i, 1])
    return position
```

The following function implements the PSO algorithm based on the previous classes and functions. There are two points worth mentioning:

First, we introduced **dynamic adjustment of the inertia weight w** by gradually reducing w during the iteration process, the global search capability can be enhanced in the early stage, and the local search can be more focused in the later stage, thereby improving the convergence effect. And because we use the rebound mechanism in boundary processing, gradually reducing the inertia weight can gradually reduce the particle speed, thereby reducing the rebound amplitude and preventing particles from bouncing out of the search space, resulting in the final result not being able to converge.

Second, we introduced **dynamic adjustment of the center point of the search space**. That is, the center of mass of the particle swarm (the mean of all particle positions) is used as the new center point of the search space.

```python
def pso_optimization(ann, X_train, y_train, dimensions, bounds, swarm_size=30, 
                     num_informants=3, max_iter=100, w_max=0.09, w_min=0.05, c1_max=2.0, c1_min=1.0, c2_max=2.0, c2_min=1.0):
    """
    PSO optimization main loop
    :param ann: neural network model
    :param X_train: training set input
    :param y_train: training set output
    :param swarm_size: particle swarm size
    :param dimensions: particle position dimensions
    :param bounds: search space range
    :param num_informants: number of informants for each particle
    :param max_iter: maximum number of iterations
    :param w: inertia weight
    :param c1: cognitive factor
    :param c2: social factor
    """
    # Initialize particle swarm
    swarm = Swarm(swarm_size, dimensions, bounds, num_informants)
    fitness_history = []  # Used to record the Global Best Fitness of each iteration
    bounds_history = []

    for iter in range(1, max_iter+1):
        # The dynamic adjustment coefficient is almost 0 in the early stage, but increases with the number of iterations and gradually approaches 1
        m = iter / max_iter

        # Dynamic Update Informants
        swarm.update_informants(num_informants, m)
        
        # Dynamically calculate the center of particle distribution
        center = np.mean([particle.position for particle in swarm.particles], axis=0)
        # Gradually reduce the search space to improve accuracy
        bounds = adaptive_bounds(bounds, m, center=center, gbest_position=swarm.gbest_position)
        bounds_range = bounds[:, 1] - bounds[:, 0] 

        # Dynamically adjust acceleration coefficient
        w = (w_max - m * (w_max - w_min)) * np.mean(bounds_range)
        
        for i, particle in enumerate(swarm.particles):
            # Calculate fitness value
            particle.fitness = fitness_function(particle.position, ann, X_train, y_train)
            
            # Update individual optimal
            if particle.fitness < particle.pbest_value:
                particle.pbest_position = particle.position.copy()
                particle.pbest_value = particle.fitness
            
            # Update Informants Optimum
            lbest_value = float('inf')
            lbest_position = np.zeros(dimensions) 
            for informant_index in swarm.informants[i]:  # Here Informants is an index
                if swarm.particles[informant_index].fitness < lbest_value:
                    lbest_value = swarm.particles[informant_index].fitness
                    lbest_position = swarm.particles[informant_index].position.copy()
            
            # Update the group optimal
            if particle.fitness < swarm.gbest_value:
                swarm.gbest_position = particle.position.copy()
                swarm.gbest_value = particle.fitness
            
            # Update particle speed and position
            update_particle(particle, lbest_position, w, c1, c2, bounds_range, m)

        fitness_history.append(swarm.gbest_value)
        bounds_history.append(bounds)
        
        if iter % 20 ==0:
            print(f"Iteration {iter}/{max_iter}, Global Best Fitness: {swarm.gbest_value}")
    
    return swarm.gbest_position, fitness_history, bounds_history
```

The following is a specific function for dynamically adjusting the search space. To increase the accuracy of the final result, we have made two changes to the search space:

First, the center of the search space will **continuously shift toward the current global optimal solution** based on the particle swarm centroid.

Second, **the search space will gradually shrink as the number of iterations increases**. However, in order to prevent particles from falling into the local optimal value too early, we have processed the shrink factor shrink_ratio to ensure that it can quickly shrink the search space in the early stage, but gradually slow down in the later stage.

```python
def adaptive_bounds(bounds, m=0, center=0, gbest_position=0, shrink_factor=1):
    """
    Dynamically adjust the search space range, center around the global optimal solution
    :param bounds: original search space range, shape (dimensions, 2)
    :param iter: current number of iterations
    :param max_iter: maximum number of iterations
    :param center: current center of particle distribution (mean of all particle positions)
    :param gbest_position: current position of global optimal solution
    :param shrink_factor: shrinking factor (controls the final shrinking range)
    :return: adjusted search space
    """
    center = np.asarray(center).flatten()
    gbest_position = np.asarray(gbest_position).flatten()
    
    # Update the center of the search space using the global optimal solution
    new_center = (center + gbest_position*m) / 2  # Comprehensive global optimal solution and particle center
    
    # Dynamically adjust the upper and lower bounds, and shrink slowly according to the square formula
    alpha = 1
    shrink_ratio = (1 - m) ** alpha
    
    min_bounds = new_center - shrink_factor * shrink_ratio * (new_center - bounds[:, 0])
    max_bounds = new_center + shrink_factor * shrink_ratio * (bounds[:, 1] - new_center)
    

    # Return the new search space range
    return np.stack([min_bounds, max_bounds], axis=1)
```

## Investigation for the balance between swarm size and iterations

```python
# Set PSO parameter combination
pso_configs = [
    {"swarm_size": 10, "max_iter": 200},  # Small population, high number of iterations
    {"swarm_size": 20, "max_iter": 100},  # Large population, few iterations
    {"swarm_size": 40, "max_iter": 50},  # Balanced distribution
    {"swarm_size": 50, "max_iter": 40},  # Balanced distribution
]

# Initialize ANN architecture (fixed architecture)
input_size = 8  # Input layer size (adjust according to your dataset)
hidden_layers = [10, 20]  # Hidden layer size
output_size = 1  # Output layer size

```

## Test for the effect of acceleration coefficients in the PSO

```python
# Set the acceleration coefficients combination
acceleration_coefficients = [
    {"c1": 1.5, "c2": 1.5},  # Balanced Combination
    {"c1": 1.5, "c2": 2.0},  # Biased towards global search
    {"c1": 2.0, "c2": 1.5},  # Biased towards local search
    {"c1": 2.0, "c2": 2.0},  # Strong search combination
]

# PSO fixed parameters
swarm_size = 30
max_iter = 100

# Fixed ANN architecture
input_size = 8  
hidden_layers = [10, 20] 
output_size = 1  
```

```python
#Store experimental results
results = []

for coeff in acceleration_coefficients:
    print(f"Testing acceleration coefficients: c1={coeff['c1']}, c2={coeff['c2']}")

    fitness_scores = []
    convergence_iterations = []
    run_times = []

    # Run the experiment multiple times to reduce the effect of randomness
    for run in range(5): 
        # Initialize ANN
        ann = NeuralNetwork(input_size=input_size, hidden_layers=hidden_layers, 
                            output_size=output_size, activation_function="relu")

        # Start timing
        start_time = time.time()

        # Optimize ANN using PSO
        dimensions = ann.to_particle().size
        bounds = ann.get_bounds()
        best_solution, fitness_history = pso_optimization(
            ann, X_train, y_train, dimensions, bounds, 
            swarm_size=swarm_size, max_iter=max_iter, 
            c1=coeff["c1"], c2=coeff["c2"]
        )

        # Stop the timer
        end_time = time.time()

        # stop T and timer
        ann.from_particle(best_solution)
        y_pred = ann.forward(X_test)
        mse = np.mean((y_pred - y_test) ** 2)
        fitness_scores.append(mse)

        # Convergence speed: detect the number of iterations to reach a certain threshold
        threshold = 350
        iter_to_threshold = next((i for i, val in enumerate(fitness_history) if val < threshold), max_iter)
        convergence_iterations.append(iter_to_threshold)

        # Record running time
        run_times.append(end_time - start_time)

    # Summarize the results
    results.append({
        "coefficients": coeff,
        "mean_fitness": mean(fitness_scores),
        "std_fitness": stdev(fitness_scores),
        "mean_convergence": mean(convergence_iterations),
        "std_convergence": stdev(convergence_iterations),
        "mean_time": mean(run_times),
        "std_time": stdev(run_times)
    })
```

```python
# Output experimental results
for result in results:
    print(f"Acceleration Coefficients: c1={result['coefficients']['c1']}, c2={result['coefficients']['c2']}")
    print(f"  Mean Fitness: {result['mean_fitness']} ± {result['std_fitness']}")
    print(f"  Mean Convergence Iterations: {result['mean_convergence']} ± {result['std_convergence']}")
    print(f"  Mean Time: {result['mean_time']} ± {result['std_time']}")
```
